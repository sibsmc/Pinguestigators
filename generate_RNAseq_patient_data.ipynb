```python
import pandas as pd
from Models import RNA_SVR
```

```python
DATA = pd.read_csv("Data/GTEx_pancreas_liver_images_liverfat_pancreasfat_seq.csv")
```

```python
def CreateDataset(kind):
    GeneNames = RNA_SVR.TissueSamples["Description"][RNA_SVR.GeneMask[kind]].values
    
    print("\n".join(GeneNames))
    
    contents = []
    
    for i, row in DATA.iterrows():
        inputs = RNA_SVR.RowToModelInput(row, kind)
        if inputs is not None:
            ds_inputs = {name: value for name, value in zip (GeneNames, inputs)}
            ds_inputs["Fat%"] = row["Fat,Percentage_%s" % kind]
            contents.append(ds_inputs)
        
        
    new_dataset = pd.DataFrame(contents, columns=["Fat%"] + list(GeneNames))
    
    # Sort by fat content, for visualization ease;
    new_dataset = new_dataset.sort_values(by=["Fat%"])
    
    
    new_dataset.to_csv("visualizeRNASeq_%s.csv" % kind)
    
    print(new_dataset.corr())
```

```python
CreateDataset("pancreas")
```

```python
CreateDataset("liver")
```
